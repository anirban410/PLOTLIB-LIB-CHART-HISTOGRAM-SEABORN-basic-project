```python
# mst plotlib Bar chat

import matplotlib.pyplot as plt
students = ["Rahul", "Madhu", "Sara", "Amit"]
marks = [85, 92, 78, 88]
plt.figure(figsize=(8, 5))
plt.bar(students, marks)
plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
# Line chart
import matplotlib.pyplot as plt
days = [1, 2, 3, 4, 5]
sales = [120, 270, 100, 280, 200]
plt.figure(figsize=(8, 5))
plt.plot(days, sales, marker='o')
plt.title("Daily Sales")
plt.xlabel("Day")
plt.ylabel("Sales")
plt.show()
```

```python
# Histogram
import matplotlib.pyplot as plt

marks = [85, 92, 78, 88, 90, 76, 84, 91, 87, 89]

plt.figure(figsize=(8, 5))
plt.hist(marks, bins=5, edgecolor='#e3fc03',)
plt.title("Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Create the dataframe
data = pd.DataFrame({
    "Department": ["AI", "AI", "ML", "DS", "DS", "AI", "ML"]
})

# Plotting the data
plt.figure(figsize=(7, 5))
sns.countplot(x="Department", data=data)
plt.title("Department Count")
plt.show()
```

```python

# Seaborn Heatmap
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

data = pd.DataFrame({
    "Marks": [85, 92, 78, 88],
    "Attendance": [90, 95, 80, 85],
    "StudyHours": [3, 5, 2, 4]
})

plt.figure(figsize=(6, 4))
sns.heatmap(data.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
